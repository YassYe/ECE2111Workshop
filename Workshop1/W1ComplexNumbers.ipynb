```python
import matplotlib.pyplot as plt
import cmath
```

# Plotting Complex Numbers

```python
# define complex number
x = 1 + 2j
x_polar = cmath.polar(x)
y_polar = cmath.polar(1+1j)
# plot the complex numbers arg(Z) = z_polar[1] , abs(Z) = z_polar[0]
plt.polar(x_polar[1], x_polar[0],marker='o')
plt.polar(y_polar[1], y_polar[0],marker='o')

plt.xticks()
plt.show()
```
